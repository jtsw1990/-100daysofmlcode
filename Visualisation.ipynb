# Matplotlib

```python
from matplotlib import pyplot as plt
from matplotlib import style # Makes graphs nice
```

```python
#style.use('seaborn-darkgrid')
```

```python
x = [5, 6, 7, 8]
y = [7, 3, 8, 3]

x2 = [5, 6, 7, 8]
y2 = [6, 7, 2, 6]

plt.plot(x, y, # Arguments after x-y is usually style sheet commands
         color = 'g', 
         linewidth = 5, label = 'line 1') 
plt.plot(x2, y2, 
         color = 'c',  
         linewidth = 10, label = 'line 2')
plt.title('ExampleGraph')
plt.ylabel('Y Axis')
plt.xlabel('X Axis')
plt.legend()
plt.grid(True, color = 'k')
plt.show()
```

```python
def duel_plot(x, y1, y2):
    
    fig = plt.figure(figsize = (15, 5))
    ax1 = fig.add_subplot(1, 2, 1)
    ax2 = fig.add_subplot(1, 2, 2)

    ax1.plot(x, y1, color = 'r')
    ax1.set_title('y1')
    ax1.set_xlabel('X Axis')
    ax1.set_ylabel('Y Axis')

    ax2.plot(x, y2)
    ax2.set_title('y2')
    ax2.set_xlabel('X Axis')
    ax2.set_ylabel('Y Axis')
    
    return fig
```

```python
duel_plot(x, y1, y2)
```

## 3D plots

```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
mpl.rcParams['legend.fontsize'] = 10
```

```python
fig = plt.figure(figsize = (10, 10))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

ax1 = fig.gca(projection = '3d')
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z ** 2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)

ax1.plot(x, y, z, label = 'parametric curve')
ax1.legend()
plt.show()
```
